```python
import matplotlib.pyplot as plt
import csv

y_real_FFT = []
y_img_FFT = []
y_comb_FFT = []

x_FFT = []

with open('FFTSpectrum.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    counter = 0
    
    firstrow = True
    for row in plots:
        
        if firstrow:
            firstrow = False
            continue
            
        y_real_FFT.append(float(row[0]))
        y_img_FFT.append(float(row[1]))
        y_comb_FFT.append(float(row[2]))

        x_FFT.append(counter)
        counter = counter + 1
```

```python
plt.plot(x_FFT,y_comb_FFT, label='FFT Spectrum')
plt.xlabel('mass')
plt.ylabel('intensity real')
plt.title('FFT Spectrum')
plt.legend()
plt.show()
```

```python
plt.plot(x_FFT,y_real_FFT, label='FFT Spectrum')
plt.xlabel('mass')
plt.ylabel('intensity real')
plt.title('FFT Spectrum')
plt.legend()
plt.show()
```

```python
plt.plot(x_FFT,y_img_FFT, label='FFT Spectrum')
plt.xlabel('mass')
plt.ylabel('intensity real')
plt.title('FFT Spectrum')
plt.legend()
plt.show()
```
